```
student_scores_high = np.random.randint(low = 80, high=100, size =(30, 3)) 
student_scores_middle = np.random.randint(low = 50, high=79, size =(20, 3)) 
student_scores_low = np.random.randint(low = 0, high=49, size =(100, 3)) 
student_scores_high
```

```
student_scores = np.insert(student_scores_high,0,student_scores_middle,axis=0)
```

```
student_scores
```

```
student_score = np.insert(student_scores, 0, student_scores_low, axis = 0)
```

```
student_score
```

```
len(student_score)
```

```
kor_score = np.array(student_score[:, 0])
kor_score
```

```
eng_score = np.array(student_score[:, 1])
eng_score
```

```
math_score = np.array(student_score[:, 2])
math_score
```

```
print('국어 최댓값 : {}'.format(np.max(kor_score)))
print('국어 최솟값 : {}'.format(np.min(kor_score)))
print('국어 평균 : {}'.format(np.mean(kor_score)))
print('국어 최빈값 : {}'.format(st.mode(kor_score)))
print('국어 중앙값 : {}'.format(np.median(kor_score)))
print('국어 분산 : {}'.format(np.var(kor_score)))
print('국어 표준편차 : {}'.format(np.std(kor_score)))
n, bins, patches = plt.hist(kor_score, bins=100)
plt.show()
```

```
print('영어 최댓값 : {}'.format(np.max(eng_score)))
print('영어 최솟값 : {}'.format(np.min(eng_score)))
print('영어 평균 : {}'.format(np.mean(eng_score)))
print('영어 최빈값 : {}'.format(st.mode(eng_score)))
print('영어 중앙값 : {}'.format(np.median(eng_score)))
print('영어 분산 : {}'.format(np.var(eng_score)))
print('영어 표준편차 : {}'.format(np.std(eng_score)))
n, bins, patches = plt.hist(eng_score, bins=100)
plt.show()
```

```
print('수학 최댓값 : {}'.format(np.max(math_score)))
print('수학 최솟값 : {}'.format(np.min(math_score)))
print('수학 평균 : {}'.format(np.mean(math_score)))
print('수학 최빈값 : {}'.format(st.mode(math_score)))
print('수학 중앙값 : {}'.format(np.median(math_score)))
print('수학 분산 : {}'.format(np.var(math_score)))
print('수학 표준편차 : {}'.format(np.std(math_score)))
n, bins, patches = plt.hist(math_score.T, bins=100)
plt.show()
```

```
# 국어, 영어, 수학 대표값은 그래프로 보듯이 중앙값으로 정할 수 있다.
# 예를 들어 직원이 100명인 회사에서 직원들 연봉 평균이 5000만원인데 사장의 연봉이 100억인 경우, 회사 전체의 연봉 평균은 1억4851만 원이 된다. 이처럼 극단적인 값이 있는 경우 중앙값이 평균값보다 유용하다.
```
