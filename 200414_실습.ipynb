```
student_score_dict = [
{'name':'김철수','kor':90, 'eng':'', 'math':0},
{'name':'박제동','kor':70, 'eng':50, 'math':100},
{'name':'홍길동','kor':80, 'eng':None, 'math':''},
{'name':'김철수','kor':90, 'eng':20, 'math':70},
{'name':'장길산','kor':None, 'eng':120, 'math':None},
{'name':'홍길동','kor':80, 'eng':None, 'math':0},
{'name':'정진수','kor':90, 'eng':0, 'math':80},
{'name':'오금순','kor':50, 'eng':70, 'math':None},
{'name':'정만득','kor':None, 'eng':None, 'math':None},
]
df_data = pd.DataFrame(student_score_dict)
df_data
```
