```
movies = ['The Holy Grail', 1975, 'The Life of Brian', 91, 'Graham cahpman',['Michael Palin', 'John Cleese', 'Terry Giliam', 'Eric Idel']]
for each_item in movies:
    print(each_item)
```

```
for movie in movies:
    if isinstance(movie, list):
        for x in movie:
            print(x)
    else:
      print(movie)
```

```
some_list = ['foo', 'bar', 'baz']
for index, value in enumerate(some_list):
    print('i:{}, v:{}'.format(index,value))
```

```
seq1 = ['foo', 'bar', 'baz']
seq2 = ['one', 'two', 'three']
seq3 = ['apple', 'raspberry', 'banana']
```

```
for a, b, c in zip(seq1, seq2, seq3):                 #zip() : 동일한 자리끼리 묶어줌
    print('a:{}, b:{}, c:{}'.format(a,b,c,))     
    print('{2}, {1}, {0}'.format(a,b,c,))   
```
