```
verb_list=[]
noun_list=[]
anonymous_list = []
count = 0

while count < 129: 
  if "다" in split_list[count]:
      verb_list.insert(count, split_list[count])
  elif (("마다" in split_list[count]) or ("또" in split_list[count]) or ("고" in split_list[count]) or ("부터" in split_list[count]) or ("을" in split_list[count]) or ("와" in split_list[count]) or ("에서" in split_list[count])
        or ("을" in split_list[count]) or ("를" in split_list[count]) or ("한" in split_list[count]) or ("한다고" in split_list[count]) or ("은" in split_list[count]) or ("에게" in split_list[count]) or ("에" in split_list[count])
        or ("는" in split_list[count]) or ("낼" in split_list[count]) or ("수" in split_list[count]) or ("과" in split_list[count]) or ("하기" in split_list[count]) or ("모든" in split_list[count]) or ("위해" in split_list[count])
        or ("로" in split_list[count]) or ("이" in split_list[count]) or ("할" in split_list[count]) or ("가" in split_list[count])):
     anonymous_list.insert(count, split_list[count])
  else:
     noun_list.insert(count, split_list[count])

  count = count + 1

print(verb_list)
print(anonymous_list)
print(noun_list)
```
