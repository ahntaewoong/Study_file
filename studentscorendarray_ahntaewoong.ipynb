```
import numpy as np
```

```
student_score =[
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 32, 85, 90, 0, 0.0, 0, 0],
[4, 42, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]

student = np.array(student_score)
```

```
student
```

```
student.shape
```

```
#각 구성원 total, avg, min, max 필드 각 생성과 값 구하기
student[:,4] = np.sum(student[:,1:4], axis = 1)
student[:,5] = np.mean(student[:,1:4], axis = 1)
student[:,6] = np.min(student[:,1:4], axis = 1)
student[:,7] = np.max(student[:,1:4], axis = 1)
```

```
student
```

```
#학급 평균 구하기
np.mean(student[:,5], axis = 0)
```

```
#과목별 평균 구하기
np.mean(student[:,1:4], axis = 0)
```
