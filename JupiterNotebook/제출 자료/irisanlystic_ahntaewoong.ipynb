## 아이리스 key 설명


### data : 각 꽃의 특성이 담겨져있다.

```
iris.data
```

### target_names : 구분해야 하는 꽃의 이름이 있다.

```
iris.target_names
```

### target : 실제로 이 수치를 가지고 있는 꽃이 무슨 꽃인가라는 정보를 정수형으로 가지고 있다. (0 - setosa / 1 - versicolor / 2 - virginica)

```
iris.target
```

### DESCR : iris 데이터에 대한 설명이다.

```
iris.DESCR
```

###feature_names : data의 특성이 무엇을 의미하는지 알려준다.

```
iris.feature_names
```

### filename : 파일의 경로를 알려준다.

```
iris.filename
```
