```
#변수 선언
students_score = [
                  {'num':'1', 'name': '김철수', 'kor' : 90, 'eng' : 80, 'math': 85},
                  {'num':'2', 'name': '박제동', 'kor' : 90, 'eng' : 85, 'math': 90},
                  {'num':'3', 'name': '홍길동', 'kor' : 80, 'eng' : 80, 'math': 80}
]
```

```
#추가
students_score = students_score + [{'num':'4', 'name': '이천웅', 'kor' : 92, 'eng' : 83, 'math': 97}, {'num':'5', 'name': '채은성', 'kor' : 91, 'eng' : 88, 'math': 92}]
```

```
def total_avg(students_score):

  for x, y in enumerate(students_score):
    y['total'] = y['kor'] + y['eng'] + y['math']
    y['avg'] = round((y['kor'] + y['eng'] + y['math']) / 3)

    print('이름 :', y['name'])
    print('총합 :', y['total'])
    print('평균 :', y['avg'])

  return students_score
```

```
student_score = total_avg(students_score)
```

```
def class_avg(c_avg):
    c_avg = students_score[0]['avg']+students_score[1]['avg']+students_score[2]['avg']+students_score[3]['avg']+students_score[4]['avg']
    print("학급평균 : ", round(c_avg/5))
    return;

class_avg(students_score)

def kor_avg(k_avg):
    k_avg = students_score[0]['kor']+students_score[1]['kor']+students_score[2]['kor']+students_score[3]['kor']+students_score[4]['kor']
    print("국어평균 : ", round(k_avg/5))
    return;

kor_avg(students_score)

def eng_avg(e_avg):
    e_avg = students_score[0]['eng']+students_score[1]['eng']+students_score[2]['eng']+students_score[3]['eng']+students_score[4]['eng']
    print("영어평균 : ", round(e_avg/5))
    return;

eng_avg(students_score)

def math_avg(m_avg):
    m_avg = students_score[0]['math']+students_score[1]['math']+students_score[2]['math']+students_score[3]['math']+students_score[4]['math']
    print("수학평균 : ", round(m_avg/5))
    return;

math_avg(students_score)
```
