```
import requests
from bs4 import BeautifulSoup
```

#CHANNEL A News

```
ur1 = 'http://www.ichannela.com/news/main/news_detailPage.do?publishId=000000198386'

r = requests.get(ur1)

soup = BeautifulSoup(r.content, 'html.parser')

Baseball = soup.find_all('div', id = 'articleSection')

print(Baseball[0].get_text())
```
