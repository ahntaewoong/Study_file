```
# 덧셈
def add(arg1, arg2):
    result = int(arg1) + int(arg2)
    print("Inside the function : ", result)
    return result
```

```
f, s = input("Input two Number? ").split(',')
Total = add(f, s)
print("Outside the function : ", Total)
```
