```
# 최대/최소 : min, max, argmin, argmax
# 통계 : sum, mean(평균), median(중간값), str(표준편차), var(분산)
```

```
x=np.array([1,2,3,1])
x.sum(), x.max(), x.argmin(), x.argmax(), x.mean(), np.median(x), x.std()
```

```
x = np.array([[0,1,2],[7,8,9],[2,3,4],[4,5,6]])
```

```
x = np.array([[1,3,5,7],[2,4,6,8]])
```

```
x, type(x), x.ndim, x.shape, len(x), x[0], x[:,3]
```

```
x.sum(axis=0), x.sum(axis=1),x[0].sum(), x[:,3].sum()  #2차원 (axis = 0 : 두 행을 더함, axis = 1 : 각 행의 열끼리만 더함)
```

```
d = np.array([[[1,3,5],[5,6,7]],[[7,8,9],[10,11,12]]])
d, d.dtype, type(d), d.ndim, d.shape, len(d), d.sum(axis = 0), d.sum(axis = 1), d.sum(axis = 2)
                                               #3차원 axis = 0 (첫번째 행렬의 첫번째 행 + 두번째 행렬의 첫번째 행)
                                               #      axis = 1 (첫번째 행렬의 합 + 두번째 행렬의 합)
                                               #      axis = 2 (각 행의 합)
```
