```
score = int(input('What is your score?'))
if score >= 90:
  letter = 'A'
elif score >= 80:
  letter = 'B' 
elif score >= 70:
  letter = 'C' 
elif score >= 60:
  letter = 'D' 
else:
  letter = 'F'
print('your score is ' + letter)
```

```
#특별 실습 변수 위치 중요!!!

n = 0
while n < 5:
  score = int(input('What is your score? '))
  n = n + 1
  if score >= 90:
    letter = 'A'
  elif score >= 80:
    letter = 'B' 
  elif score >= 70:
    letter = 'C' 
  elif score >= 60:
    letter = 'D' 
  else:
    letter = 'F'
  print(n, 'your score is ' + letter)
  
```
