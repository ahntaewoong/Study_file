```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```
plt.rcParams['figure.figsize'] = [5,5]
```

```
season_2017 = pd.read_excel('./sample_data/2017년 KBO순위.xlsx')
season_2017
```

```
R_2017 = season_2017['득점']
R_2017
```

```
r_2017 = season_2017['실점']
r_2017
```

```
PE_2017 = R_2017**2/(R_2017**2+r_2017**2)
PE_2017
```

```
season_2017['PE승률'] = round(PE_2017,3)
season_2017
```

```
D_2017 = season_2017['승률']-season_2017['PE승률']
D_2017
```

```
season_2017['승률차이'] = D_2017
season_2017
```

```
season_2017['불펜WAR'] = np.array([4.07,9.09,10.21,11.07,4.11,4.37,5.65,6.58,6.04,6.46])
season_2017
```
