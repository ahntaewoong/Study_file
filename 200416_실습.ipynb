###html

```
html = "<html> <body> \
<h1 class='public_class_name' id='h1_id_name'>[1]크롤링이란?</h1> \
<p class='public_class_name' id='p01_id_name'>웹페이지에서 필요한 데이터를 추출하는 것</p> \
<p id='p02_id_name' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p> \
 </body> </html>"
```

```
soup = BeautifulSoup(html, 'html.parser')
```

```
title_data = soup.find('h1')        #tag로 검색 
```

```
type(title_data), title_data, title_data.string
```

```
title_data = soup.find_all(id = 'h1_id_name')    #id로 검색
```

```
title_data
```

```
title_data[0].get_text()
```
