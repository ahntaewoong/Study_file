### data : 각 병의 특성이 담겨져있다.

```
data.data
```

### DESCR : Cancer 데이터에 대한 설명이다.

```
data.DESCR
```

### feature_names : data의 특성이 무엇을 의미하는지 알려준다.

```
data.feature_names
```

### target_names : 악성, 양성

```
data.target_names
```

### target : 실제로 이 데이터에 대해서 악성인지 양성인가라는 정보를 정수형으로 가지고 있다. (0 - malignant(악성종양) / 1 - benign(양성종양)

```
data.target
```

### filename : 파일의 경로를 알려준다.


```
data.filename
```
